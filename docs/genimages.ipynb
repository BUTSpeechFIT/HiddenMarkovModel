```julia
using Pkg
Pkg.activate("../")

using MarkovModels
```

```julia
SF = LogSemifield{Float64}

fsm = FSM{SF}()

s1 = addstate!(fsm, pdfindex = 1, label = "a" )
s2 = addstate!(fsm, pdfindex = 2, label = "b" )
link!(fsm, s1, s2)
link!(fsm, s2, s2)
link!(fsm, s2, s1)

setinit!(s1)
setfinal!(s2)
fsm |> renormalize!
```

```julia
fsm = FSM{SF}()

s1 = addstate!(fsm, pdfindex = 1)
s2 = addstate!(fsm, pdfindex = 2, label = "a")
s3 = addstate!(fsm, label = "b")
s4 = addstate!(fsm)

fsm
```

```julia
setinit!(s1)
setfinal!(s4)

fsm
```

```julia
link!(fsm, s1, s1, SF(log(1/2)))
link!(fsm, s1, s2, SF(log(1/2)))
link!(fsm, s2, s3)
link!(fsm, s3, s4)
fsm
```
